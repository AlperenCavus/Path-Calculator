```python
def minCostPathBacktrack(grid):
    m, n = len(grid), len(grid[0]) #m by n matrix
    min_cost = float('inf')  # Initialize minimum cost to infinity
    min_path = []

    def backtrack(i, j, pathSum, path):
        nonlocal min_cost, min_path

        # Prune if cumulative cost exceeds minimum cost
        if pathSum >= min_cost:  # Prune if cumulative cost exceeds minimum cost
            return

        if i == m - 1 and j == n - 1:  # if we are at bottom left
            min_cost = pathSum
            min_path = path
            return

        if i == m - 1:  # if we are at bottom row
            backtrack(i, j + 1, pathSum + grid[i][j + 1], path + [(i, j + 1)])
            return

        if j == n - 1:  # if we are at rightmost column
            backtrack(i + 1, j, pathSum + grid[i + 1][j], path + [(i + 1, j)])
            return

        # Explore both right and down paths
        backtrack(i, j + 1, pathSum + grid[i][j + 1], path + [(i, j + 1)])
        backtrack(i + 1, j, pathSum + grid[i + 1][j], path + [(i + 1, j)])

    # Start exploration from (0, 0)
    backtrack(0, 0, grid[0][0], [(0, 0)])

    return min_cost, min_path

# Example test cases
grid1 = [[1,3,1],
         [1,10,1],
         [10,2,1]]

grid2 = [[1,2,3],
         [4,5,6],
         [7,8,9]]

grid3 = [[4,2,3,5,1],
         [2,3,1,5,4],
         [5,1,2,4,3],
         [3,5,4,1,2],
         [1,4,5,2,3]]

grid4 = [
    [4, 2, 3, 5, 1, 9, 3, 6, 7, 10],
    [2, 3, 1, 5, 4, 10, 9, 8, 7, 3],
    [5, 1, 2, 4, 3, 10, 9, 8, 7, 2],
    [3, 5, 4, 1, 2, 9, 10, 8, 7, 4],
    [1, 4, 5, 2, 3, 9, 10, 8, 7, 5],
    [3, 5, 4, 1, 2, 9, 10, 8, 7, 8],
    [2, 3, 1, 5, 4, 10, 9, 8, 7, 9],
    [5, 1, 2, 4, 3, 10, 9, 8, 7, 1],
    [3, 5, 4, 1, 2, 9, 10, 8, 7, 1],
    [1, 4, 5, 2, 3, 9, 10, 8, 7, 2]
]

print(minCostPathBacktrack(grid1)[0]) # Expected output: 7
print(minCostPathBacktrack(grid1)[1])
print('-------------')
print(minCostPathBacktrack(grid2)[0]) # Expected output: 21
print(minCostPathBacktrack(grid2)[1])
print('-------------')
print(minCostPathBacktrack(grid3)[0])
print(minCostPathBacktrack(grid3)[1])
print('-------------')
print(minCostPathBacktrack(grid4)[0])
print(minCostPathBacktrack(grid4)[1])
```
